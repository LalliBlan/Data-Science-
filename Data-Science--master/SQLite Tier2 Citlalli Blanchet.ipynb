```python
#Question 12:
#/* Q12: Find the facilities with their usage by member, but not guests */

import sqlite3
from sqlite3 import Error
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
SELECT f.name,COUNT(m.surname) AS numMember
FROM Bookings AS b
 LEFT JOIN 
     Members AS m
     ON b.memid=m.memid 
LEFT JOIN 
     Facilities AS f
     ON b.facid=f.facid
WHERE m.memid <> 0
GROUP BY f.name;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "/Users/jose1/Desktop/SpringboardLalli/SQLFiles_Tier_1/sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#Question 13:
#/* Q13: Find the facilities usage by month, but not guests */

import sqlite3
from sqlite3 import Error
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """        
SELECT f.name, COUNT(f.facid) AS NumMembers,month
FROM Facilities AS f
LEFT JOIN (
SELECT b.starttime,strftime('%m',b.starttime) AS month,m.memid,b.facid
FROM Bookings AS b
LEFT JOIN 
     Members AS m
     ON b.memid=m.memid
WHERE m.memid <> 0) AS s
ON f.facid=s.facid
ORDER BY f.name,month;
"""
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "/Users/jose1/Desktop/SpringboardLalli/SQLFiles_Tier_1/sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
